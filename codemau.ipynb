```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from scipy.signal import convolve2d
#Dùng để áp dụng phép tích chập 2D , làm mờ ảnh 
```

```python
image = cv2.imread('hinh2.jpg')

# Chuyển ảnh từ BGR (OpenCV) sang RGB (Matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Hiển thị ảnh
plt.imshow(image_rgb)
plt.axis('off')  # Tắt trục toạ độ
plt.title( 'Ảnh gốc')
plt.show()
```

```python
gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Chuyển sang ảnh xám ")
plt.show()
```

```python
K = np.ones((10, 15)) / (10 * 15)
Out = convolve2d(gray, K, boundary='symm', mode='same') #Kết quả đầu ra có cùng kích thước với ảnh đầu vào
plt.imshow(Out, cmap='gray')
plt.title("Ảnh sau khi làm mờ")
plt.show()
```

```python
threshold = 150
binaryimg = Out > threshold
plt.imshow(binaryimg, cmap='gray')
plt.title("Ảnh nhị phân sau khi tách ngưỡng")
plt.show()
```
